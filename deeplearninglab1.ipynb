```python
import skimage as ski
from skimage.transform import resize

image = ski.io.imread('/content/drive/MyDrive/download.jpeg')
# ... or any other NumPy array!
edges = ski.filters.sobel(image)
r = resize(edges, (128, 128))
ski.io.imshow(r)
ski.io.show()

blur = ski.filters.gaussian(image, sigma=1)
p = resize(blur, (128, 128))
ski.io.imshow(p)
ski.io.show()

median= ski.filters.median(image)
o = resize(median, (128, 128))
ski.io.imshow(o)
ski.io.show()

lap = ski.filters.laplace(image)
i = resize(lap, (128, 128))
ski.io.imshow(i)
ski.io.show()

eschar = ski.filters.scharr(image)
y = resize(eschar, (128, 128))
ski.io.imshow(y)
ski.io.show()

from skimage import data, img_as_float
from skimage.color import rgb2gray
img = img_as_float(image)
gray= rgb2gray(img)
rob = ski.filters.roberts(gray)
u = resize(rob, (128, 128))
ski.io.imshow(u)
ski.io.show
```
